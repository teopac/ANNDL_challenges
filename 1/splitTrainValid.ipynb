# Creation of Dataset
This script splits the images in the training folder in the training and valdation set, saving them divided by class.
The user has to express the percentage of images that will be put in the validation set.
This script has to be placed in the same directory of the training and test folders.



```python
import json
import os
import shutil
import random
```

```python
jsonFile = open("train_gt.json", "r")
a = json.load(jsonFile)
#params
SEED = "1234"
percentage = 10 #esprimere come intero plz
random.seed(SEED)
```

```python
#important directories
currDir =  os.getcwd()
trainOrDir = os.path.join(currDir, "training")
trainDir = os.path.join(currDir, "train")
validationDir = os.path.join(currDir, "validation")
testDir = os.path.join(currDir, "test")
```

```python
#classes of our problem
classes = ["noPerson", "allPeople", "someOne"]
```

```python
#creation of the folders
os.mkdir(trainDir)
os.mkdir(validationDir)
for label in classes:
    os.mkdir(os.path.join(trainDir, label))
    os.mkdir(os.path.join(validationDir, label))
```

```python
#images are filtered wrt the classes they belong
noPersonImages=[key.encode("utf-8") for key in a.keys() if int(a[key]) is 0]
allPeopleImages=[key.encode("utf-8") for key in a.keys() if a[key] == 1]
someOneImages=[key.encode("utf-8") for key in a.keys() if a[key] == 2]
```

```python
#images are shuffled in order to avoid explicitordering on the dataset
random.shuffle(noPersonImages)
random.shuffle(allPeopleImages)
random.shuffle(someOneImages)
print(len(allPeopleImages)/10)
```

```python
#images are divided between training nad validation and copied in the target directory

#no mask images
print("copio immagini senza mascherina")
for image in noPersonImages[:len(noPersonImages)/percentage]:
    fileName = os.path.join(trainOrDir, image.encode("utf-8"))
    destination = os.path.join(validationDir, classes[a[image]])
    shutil.copy(fileName, destination)
for image in noPersonImages[len(noPersonImages)/percentage:]:
    fileName = os.path.join(trainOrDir, image)
    destination = os.path.join(trainDir, classes[a[image]])
    shutil.copy(fileName, destination)
    
# all mask images    
print("copio immagini tutti con mascherina")    
for image in allPeopleImages[:len(allPeopleImages)/percentage]:
    fileName = os.path.join(trainOrDir, image)
    destination = os.path.join(validationDir, classes[a[image]])
    shutil.copy(fileName, destination)
for image in allPeopleImages[len(allPeopleImages)/percentage:]:
    fileName = os.path.join(trainOrDir, image)
    destination = os.path.join(trainDir, classes[a[image]])
    shutil.copy(fileName, destination)  

#some mas images    
print("copio immagini alcuni con mascherina")
for image in someOneImages[:len(someOneImages)/percentage]:
    fileName = os.path.join(trainOrDir, image)
    destination = os.path.join(validationDir, classes[a[image]])
    shutil.copy(fileName, destination)
for image in someOneImages[len(someOneImages)/percentage:]:
    fileName = os.path.join(trainOrDir, image)
    destination = os.path.join(trainDir, classes[a[image]])
    shutil.copy(fileName, destination)    
```

# BAGGING TEST RESULTS
This code takes the results of the models in the Result folder, aggregates them and applies majority voting. The models that are bagged have already been selected during validation.

```python
import os
import csv
import numpy as np
from scipy import stats
```

```python
#path with the predictions on the test set of the models that have been chosen
results_path = "Results"
```

```python
os.listdir(results_path)
```

```python
#reomves a folder that sometimes is saved in the Result folder
content = os.listdir(results_path)
content.remove('.ipynb_checkpoints')
content
```

```python
#load from the csv the prediction on the test set of each model
models = []
for result in content:
    res_path = os.path.join(results_path, result)
    f = open(res_path)
    reader = csv.DictReader(f, delimiter=",")
    a = {}
    for row in reader:
        a[row["Id"]] = int(row["Category"]) 
    f.close()   
    models.append(a)
```

```python
#creation of the dictionary that will hold the result of the bagging
listDict = {}
for key in models[0].keys():
    listDict[key] = []

for model in models:
    #sav for every image the vote of the current model
    for key in model.keys():
        listDict[key].append(model[key])
```

```python
#apply the mode to the lists of note in order to do majority voting
for key in listDict:
    array = listDict[key]
    mode = stats.mode(array).mode[0]
    listDict[key] = mode
```

```python
#write the results in a csv
with open("bagged_results.csv", "w") as r:
    writer = csv.DictWriter(r, delimiter=",", fieldnames = ["Id", "Category"] )
    writer.writeheader()
    for key in listDict:
        writer.writerow({"Id":key, "Category":listDict[key]})
        
    
```
